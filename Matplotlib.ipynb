**Matplotlib**

-> Matplotlib is a powerful plotting library in Python used for creating static, animated, and interactive visualizations.

-> Matplotlibâ€™s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Basic Graph**

```
# line 1
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Keyword Argument Notation
plt.plot(x,y, label = '2x',color='blue',linewidth=1,linestyle='--',marker='*',
         markersize=7,markeredgecolor='red')

# line 2
# x2 = [0,1,2,3,4]
y2 = [i**2 for i in x]

# plt.plot(x2,y2,'r*-',label='x^2')

# Plot part of the graph as line
plt.plot(x[:3],y2[:3], 'r*-',label = 'x^2')
# Plot remainder of graph as a dot
plt.plot(x[2:],y2[2:], 'r*--')

# Add a title (specify font parameters with fontdict)
plt.title('My First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

# Add annotations for specific points
plt.annotate('Point 1', xy=(1, 2), xytext=(1, 3),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Point 2', xy=(2, 4), xytext=(2, 5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Point 3', xy=(3, 6), xytext=(3, 7),
             arrowprops=dict(facecolor='black', shrink=0.05))

# plt.fill_between(x,y,y2,color='green',aplha=0.5)
# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

```
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line chart
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

# Add title and labels
plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

```
import matplotlib.pyplot as plt
import random as random

plt.figure(figsize=(8, 6))
students = ["Jane","Joe","Beck","Tom",
            "Sam","Eva"]

marks=[]
for i in range(0,len(students)):
     marks.append(random.randint(0, 101))


plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("CLASS RECORDS")
plt.plot(students,marks,'b--',marker='*')
plt.show()
```

**Bar Chart**

```
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values,color='maroon',width=0.7)
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    # bar.set_hatch(patterns.pop())

plt.savefig('barchart.png', dpi=300)

plt.show()
```

```
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Set position of bar on X axis
br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# print(br1)
# print(br2)
# print(br3)


# Make the plot
patterns= ['/', 'O', '*','.','|']
x = plt.bar(br1,IT, color ='r', width = barWidth,
        edgecolor ='grey', label ='IT')

for bar in x:
    bar.set_hatch(patterns.pop(0))

patterns= ['/', 'O', '*','.','|']

y = plt.bar(br2,ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')

for bar in y:
    bar.set_hatch(patterns.pop(0))

patterns= ['/', 'O', '*','.','|']

z = plt.bar(br3,CSE, color ='b', width = barWidth,
        edgecolor ='grey', label ='CSE')

for bar in z:
    bar.set_hatch(patterns.pop(0))

# Adding labels and title
plt.xlabel('Branch', fontweight ='bold', fontsize = 10)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 10)
plt.title('Comparison of IT, ECE, and CSE')

#adding xticks
plt.xticks([r + barWidth for r in range(len(IT))],[2015,2016,2017,2018,2019])

plt.legend()
plt.show()
```

```
#Horizontal bar graph
y=['one', 'two', 'three', 'four', 'five']

# getting values against each value of y
x=[5,24,35,67,12]
plt.barh(y, x)

# setting label of y-axis
plt.ylabel("pen sold")

# setting label of x-axis
plt.xlabel("Quantity")
plt.title("Horizontal bar graph")
plt.show()
```

```
#stacked bar graph
# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```

```
d = {'Fruits':['Apple','Mango','Banana','Orange'],
     'Colour':['Red','Yellow','Yellow','Orange']}
bins=[0,1,2,3]
df = pd.DataFrame(d,index=['Prod1','Prod2','Prod3','Prod4'])
plt.hist(df['Colour'],bins=bins,color='red',edgecolor='black',linewidth=1)
plt.title('Fruits and it\'s Colour')
plt.xlabel('Colour')
plt.ylabel('Count')
plt.xticks(bins)
plt.show()
```
