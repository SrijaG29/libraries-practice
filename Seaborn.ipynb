**Seaborn**

```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
sns.get_dataset_names()
```

```
tips = sns.load_dataset("tips")
tips
```

We can plot all columns at a time using seaborn

```
plt.figure(figsize=(16,16))
sns.lineplot(tips)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

```
plt.figure(figsize = (15,15))
sns.lineplot(tips["total_bill"])
sns.lineplot(tips["tip"])
plt.show()
```

**Histogram**

Returns values and its count

```
sns.histplot(tips['tip'])
```

```
sns.histplot(tips['tip'],kde=True) #kde is used to draw wave with points.
```

**Bar plot**

```
sns.barplot(x='sex',y='tip',data=tips,hue='day')
#hue is used to group the data according to the column.
```

```
sns.barplot(x='day',y='tip',data=tips,hue='sex')
```

**Strip plot**

```
sns.stripplot(x = 'day',y = 'tip',data = tips,hue='sex')
```

```
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind='reg')
```

```
sns.pairplot(titanic.select_dtypes(include='number'))
```
